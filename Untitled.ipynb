```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
dfs = pd.read_csv('parkinsons/parkinsons.data')
print(dfs.shape)
print(dfs.info())
dfs.head()
```

```python
dfs.describe()
```

```python
#Find missing values in our dataset:
dfs.isnull().sum()
```

```python
#Check if we have any duplicated data
dfs.duplicated().sum()
```

```python
#It appears there is no column with a missing value, and no duplicated data.
#We can get rid of the name column before plotting our data, as it is not relevant
dfs.drop(columns='name', inplace = True)
```

```python
corr_matrix = dfs.corr().abs()
fig, ax = plt.subplots(figsize=(20,20))         # Sample figsize in inches

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)
plt.show()
```

```python
#Split the data into features and labels
X = dfs.drop(columns='status')
vif = pd.DataFrame()
vif["feature"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif.sort_values('VIF', ascending=False)
y = dfs['status']
```

```python
#Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size = .8)
```
