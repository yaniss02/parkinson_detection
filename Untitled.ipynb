```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score









```

```python
#Reading and Loading the data
dfs = pd.read_csv('parkinsons/parkinsons.data')

```

```python
#Getting basic info about the data
print(dfs.shape)
print(dfs.info())
dfs.head()
```

```python
#Getting basic stats about the data
dfs.describe()
```

```python
#Find missing values in our dataset:
dfs.isnull().sum()
```

```python
#Check if we have any duplicated data
dfs.duplicated().sum()
```

```python
#It appears there is no column with a missing value, and no duplicated data.
#We can get rid of the name column before plotting our data, as it is not relevant
dfs.drop(columns='name', inplace = True)
```

```python
#Engineer new feature for the frequency range:
dfs['range'] = dfs['MDVP:Fhi(Hz)'] - dfs['MDVP:Flo(Hz)']
print(dfs.head())
```

```python
dfs.hist(figsize=(25,20))
```

```python
corr_matrix = dfs.corr().abs()
fig, ax = plt.subplots(figsize=(20,20))         # Sample figsize in inches
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)
plt.show()
```

```python
# Extract upper triangle to avoid duplicate values
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Count pairs with correlation > 0.9
num_high_corr_90 = (upper_triangle > 0.9).sum().sum()

# Count pairs with correlation > 0.85
num_high_corr_85 = (upper_triangle > 0.85).sum().sum()

print(f"Number of feature pairs with correlation > 0.9: {num_high_corr_90}")
print(f"Number of feature pairs with correlation > 0.85: {num_high_corr_85}")
```

```python
#Finding feature importance
if 'status' in dfs.columns:
    #Separating the dataset between the features and the target
    X = dfs.drop(columns=['status'])  # Features
    y = dfs['status']  # Target variable

    # Fit a Random Forest model
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X, y)

    # Extract feature importances
    feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': rf.feature_importances_})
    feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

    # Display feature importance
    print(feature_importance)

else:
    print("Target variable 'status' is missing. Cannot compute feature importance.")
```

```python

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Get feature importance
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': model.feature_importances_})
feature_importance = feature_importance.set_index('Feature')

# Step 3: Identify and remove the less important feature in each correlated pair
features_to_drop = set()

for column in upper_triangle.columns:
    for index in upper_triangle.index:
        if upper_triangle.loc[index, column] > 0.9:
            # Compare feature importance scores
            if feature_importance.loc[index, 'Importance'] > feature_importance.loc[column, 'Importance']:
                features_to_drop.add(column)  # Drop the less important feature
            else:
                features_to_drop.add(index)

# Drop the selected features
df_reduced = dfs.drop(columns=features_to_drop)
print(df_reduced.head())

```

```python
#New df, for testing, without skewed values:
skewed_df=df_reduced
skewed_df = skewed_df[skewed_df["MDVP:Fhi(Hz)"] < 300]
skewed_df = skewed_df[skewed_df['MDVP:Jitter(Abs)'] < 0.00013]
skewed_df = skewed_df[skewed_df['MDVP:APQ'] < 0.07]
skewed_df = skewed_df[skewed_df['range'] < 250]
skewed_df = skewed_df[skewed_df['DFA'] > 0.6250]
skewed_df = skewed_df[skewed_df['PPE'] < 0.4]



#skewed_df = skewed_df[skewed_df['Jitter:DDP'] < 0.04]


skewed_df.skew()
```

```python
skewed_df.shape
X = df_reduced.drop(columns='status')
y = df_reduced['status']
print(X.shape)
print(y.shape)
```

```python
#Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size = .8)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Logistic Regression.
# Decision Trees
# Random Forest
# Support Vector Machine (SVM)
```

```python
logistic_model = LogisticRegression(random_state=42)
logistic_model.fit(X_train, y_train)
logistic_preds = logistic_model.predict(X_test)
print("Logistic Regression Classification Report:")
print(classification_report(y_test, logistic_preds))
```

```python
logistic_accuracy = accuracy_score(y_test, logistic_preds)
print(f"Logistic Regression Accuracy: {logistic_accuracy:.2f}")
```

```python
pd.DataFrame({'actual':y_test,'predict':logistic_preds})
```

```python
cm=confusion_matrix(y_test,logistic_preds)
plt.figure(figsize=(8,6))
fg=sns.heatmap(cm,annot=True,cmap="Blues")
figure=fg.get_figure()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Output Confusion Matrix")
```

```python
tree_model = DecisionTreeClassifier(random_state=42,max_depth=5)
tree_model.fit(X_train, y_train) 
tree_preds = tree_model.predict(X_test)
print("Decision Tree Classification Report:")
print(classification_report(y_test, tree_preds))
```

```python
tree_accuracy = accuracy_score(y_test, tree_preds)
print(f"Decision Tree Accuracy: {tree_accuracy:.2f}")
```

```python
pd.DataFrame({'actual':y_test,'predict':tree_preds})
```

```python
cm=confusion_matrix(y_test,tree_preds)
plt.figure(figsize=(8,6))
fg=sns.heatmap(cm,annot=True,cmap="Blues")
figure=fg.get_figure()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Output Confusion Matrix")
```

```python
# Random Forest
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train) 
rf_preds = rf_model.predict(X_test)
print("Random Forest Classification Report:")
print(classification_report(y_test, rf_preds))
```

```python
rf_accuracy = accuracy_score(y_test, rf_preds)
print(f"Decision Tree Accuracy: {rf_accuracy:.2f}")
```

```python
pd.DataFrame({'actual':y_test,'predict':rf_preds})
```

```python
cm=confusion_matrix(y_test,rf_preds)
plt.figure(figsize=(8,6))
fg=sns.heatmap(cm,annot=True,cmap="Blues")
figure=fg.get_figure()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Output Confusion Matrix")
```

```python
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(X_train, y_train)
svm_predictions = svm_model.predict(X_test)
print("SVM Classification Report:")
print(classification_report(y_test, svm_predictions))
```

```python
svm_accuracy = accuracy_score(y_test, svm_predictions)
print(f"Decision Tree Accuracy: {svm_accuracy:.2f}")
```

```python
pd.DataFrame({'actual':y_test,'predict':svm_predictions})
```

```python
cm=confusion_matrix(y_test,svm_predictions)
plt.figure(figsize=(8,6))
fg=sns.heatmap(cm,annot=True,cmap="Blues")
figure=fg.get_figure()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Output Confusion Matrix")
```

```python
ensemble_model = VotingClassifier(estimators=[
    ('logistic', logistic_model),
    ('tree', tree_model),
    ('rf', rf_model),
    ('svm', svm_model)
], voting='hard')
ensemble_model.fit(X_train, y_train)
ensemble_preds = ensemble_model.predict(X_test)
print("Ensemble Model Classification Report:")
print(classification_report(y_test, ensemble_preds))
```

```python
ensemble_accuracy = accuracy_score(y_test, ensemble_preds)
print(f"Ensemble Model Accuracy: {ensemble_accuracy:.2f}")
```

```python
pd.DataFrame({'actual':y_test,'predict':ensemble_preds})
```

```python
cm =confusion_matrix(y_test,ensemble_preds)
plt.figure(figsize=(8,6))
fg=sns.heatmap(cm,annot=True,cmap="Blues")
figure=fg.get_figure()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Output Confusion Matrix")
plt.show()
```

```python
model_preds = {
    'Logistic Regression': logistic_preds,
    'Random Forest': rf_preds,
    'SVM': svm_predictions,
    'Decision Trees': tree_preds,
    'Ensemble Models': ensemble_preds  # Enable probability for AUC-ROC
}
results = {'Model': [], 'Accuracy': [], 'AUC-ROC': [], 'Precision': [], 'Recall': [], 'F1-Score': [], 'AUC-ROC': []}

for model_name, y_pred in model_preds.items():
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
    auc_roc = roc_auc_score(y_test, y_pred)
    
    # Append results to the dictionary
    results['Model'].append(model_name)
    results['Accuracy'].append(accuracy)
    results['Precision'].append(precision)
    results['Recall'].append(recall)
    results['F1-Score'].append(f1)
    results['AUC-ROC'].append(auc_roc)

# Convert results to NumPy array for plotting
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC-ROC']
values = np.array([results[m] for m in metrics])

# Plot the results
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
sns.set_palette("ch:s=.25,rot=-.25")

x = np.arange(len(model_preds))  # X positions for bars
width = 0.15  # Width of bars

# Plot each metric
for i, metric in enumerate(metrics):
    plt.bar(x + i * width, values[i], width, label=metric)

plt.xticks(x + width, results['Model'], rotation=30, ha='right')
plt.ylabel("Score")
plt.title("Model Performance Comparison")
plt.legend()
plt.show()

# Print the metrics for each model
for model_name in results['Model']:
    print(f"{model_name}:")
    for metric in metrics:
        print(f"  {metric}: {results[metric][results['Model'].index(model_name)]:.2f}")
```
