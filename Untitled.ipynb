```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.ensemble import RandomForestClassifier

```

```python
#Reading and Loading the data
dfs = pd.read_csv('parkinsons/parkinsons.data')

```

```python
#Getting basic info about the data
print(dfs.shape)
print(dfs.info())
dfs.head()
```

```python
#Getting basic stats about the data
dfs.describe()
```

```python
#Find missing values in our dataset:
dfs.isnull().sum()
```

```python
#Check if we have any duplicated data
dfs.duplicated().sum()
```

```python
#It appears there is no column with a missing value, and no duplicated data.
#We can get rid of the name column before plotting our data, as it is not relevant
dfs.drop(columns='name', inplace = True)
```

```python
#Engineer new feature for the frequency range:
dfs['range'] = dfs['MDVP:Fhi(Hz)'] - dfs['MDVP:Flo(Hz)']
print(dfs.head())
```

```python
corr_matrix = dfs.corr().abs()
fig, ax = plt.subplots(figsize=(20,20))         # Sample figsize in inches
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax)
plt.show()
```

```python
# Extract upper triangle to avoid duplicate values
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Count pairs with correlation > 0.9
num_high_corr_90 = (upper_triangle > 0.9).sum().sum()

# Count pairs with correlation > 0.85
num_high_corr_85 = (upper_triangle > 0.85).sum().sum()

print(f"Number of feature pairs with correlation > 0.9: {num_high_corr_90}")
print(f"Number of feature pairs with correlation > 0.85: {num_high_corr_85}")
```

```python
#Finding feature importance
if 'status' in dfs.columns:
    #Separating the dataset between the features and the target
    X = dfs.drop(columns=['status'])  # Features
    y = dfs['status']  # Target variable

    # Fit a Random Forest model
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X, y)

    # Extract feature importances
    feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': rf.feature_importances_})
    feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

    # Display feature importance
    print(feature_importance)

else:
    print("Target variable 'status' is missing. Cannot compute feature importance.")
```

```python

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Get feature importance
feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': model.feature_importances_})
feature_importance = feature_importance.set_index('Feature')

# Step 3: Identify and remove the less important feature in each correlated pair
features_to_drop = set()

for column in upper_triangle.columns:
    for index in upper_triangle.index:
        if upper_triangle.loc[index, column] > 0.9:
            # Compare feature importance scores
            if feature_importance.loc[index, 'Importance'] > feature_importance.loc[column, 'Importance']:
                features_to_drop.add(column)  # Drop the less important feature
            else:
                features_to_drop.add(index)

# Drop the selected features
df_reduced = dfs.drop(columns=features_to_drop)
print(df_reduced.head())

```

```python
corr_matrix_important_features = df_reduced.corr().abs()
fig, ax = plt.subplots(figsize=(20,20))         # Sample figsize in inches

sns.heatmap(corr_matrix_important_features, annot=True, cmap='coolwarm', ax=ax)
plt.show()

```
